```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
np.random.seed(42)
```

```python
data = datasets.make_blobs(
    n_samples=500,
    n_features=2,
    centers=5,
    cluster_std=5,
    center_box=(20, 90),
    shuffle=True,
    random_state=42
)
```

```python
plt.figure()
plt.scatter(data[0][:, 0], data[0][:, 1])
plt.show()
```

```python
df = pd.DataFrame(data[0], columns=['x', 'y'])
```

```python
df.to_csv('./blob.csv', sep=',', encoding='utf-8')
```
